### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2 = pd.read_csv(url2)

tienda2.head()
```

```python
tienda3 = pd.read_csv(url3)

tienda3.head()
```

```python
tienda4 = pd.read_csv(url4)

tienda4.head()
```

#1. Análisis de facturación



```python
#Venta en total de cada tienda, se sumo todas las ventas de cada categoría, diferenciados por tiendas

total_venta_tienda = tienda["Precio"].sum()
total_venta_tienda2 = tienda2["Precio"].sum()
total_venta_tienda3 = tienda3["Precio"].sum()
total_venta_tienda4 = tienda4["Precio"].sum()
print(total_venta_tienda)
print(total_venta_tienda2)
print(total_venta_tienda3)
print(total_venta_tienda4)
```

# 2. Ventas por categoría

```python
print("Tienda 1")
for categoria in (tienda["Categoría del Producto"].unique()):
  print("-", categoria)

print("Tienda 2")
for categoria in (tienda2["Categoría del Producto"].unique()):
  print("-", categoria)

print("Tienda 3")
for categoria in (tienda3["Categoría del Producto"].unique()):
  print("-", categoria)

print("Tienda 4")
for categoria in (tienda4["Categoría del Producto"].unique()):
  print("-", categoria)
```

```python
#Tienda 1

def suma_categoria(categoria):
  suma_categoria = tienda[tienda["Categoría del Producto"]== categoria ]
  return suma_categoria["Precio"].sum()

for categoria in (tienda["Categoría del Producto"].unique()):
  print(categoria, suma_categoria(categoria))
```

```python
#Tienda 2
def suma_categoria(categoria):
  suma_categoria = tienda2[tienda2["Categoría del Producto"]== categoria ]
  return suma_categoria["Precio"].sum()

for categoria in (tienda2["Categoría del Producto"].unique()):
  print(categoria, suma_categoria(categoria))
```

```python
#Tienda 3

def suma_categoria(categoria):
  suma_categoria = tienda3[tienda3["Categoría del Producto"]== categoria ]
  return suma_categoria["Precio"].sum()

for categoria in (tienda3["Categoría del Producto"].unique()):
  print(categoria, suma_categoria(categoria))
```

```python
#Tienda 4

def suma_categoria(categoria):
  suma_categoria = tienda4[tienda4["Categoría del Producto"]== categoria ]
  return suma_categoria["Precio"].sum()

for categoria in (tienda4["Categoría del Producto"].unique()):
  print(categoria, suma_categoria(categoria))
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = tienda["Producto"].value_counts()
print(productos_vendidos)
```

```python
print("🔝 Productos más vendidos:")
print(productos_vendidos.head())

print("\n🔻 Productos menos vendidos:")
print(productos_vendidos.tail())
```

```python
productos_vendidos2 = tienda2["Producto"].value_counts()
print("🔝 Productos más vendidos:")
print(productos_vendidos2.head())

print("\n🔻 Productos menos vendidos:")
print(productos_vendidos2.tail())
```

```python
productos_vendidos3 = tienda3["Producto"].value_counts()
print("🔝 Productos más vendidos:")
print(productos_vendidos3.head())

print("\n🔻 Productos menos vendidos:")
print(productos_vendidos3.tail())
```

```python
productos_vendidos4 = tienda4["Producto"].value_counts()
print("🔝 Productos más vendidos:")
print(productos_vendidos4.head())

print("\n🔻 Productos menos vendidos:")
print(productos_vendidos4.tail())
```

# 5. Envío promedio por tienda

```python
tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista con tus DataFrames
for i, df in enumerate(tiendas, start=1):
    promedio = df["Costo de envío"].mean()
    print(f"📦 Envío promedio tienda {i}: {promedio:.2f}")
```
